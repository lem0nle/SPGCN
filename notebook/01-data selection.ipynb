```python
# 导出成 dgl 图

import dgl
import torch as th

edges = []
for k, cs in comp_gk.items():
    u = comp_ind[k]
    for c in cs:
        v = comp_ind[c]
        edges.append((u, v))

for k, cs in comp_wgk.items():
    u = comp_ind[k]
    for c in cs:
        v = comp_ind[c]
        edges.append((u, v))
        
u, v = zip(*edges)
u, v = th.tensor(u), th.tensor(v)
g = dgl.graph((u, v))
```
